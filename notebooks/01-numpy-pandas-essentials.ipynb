## NumPy: La base del cálculo numérico

NumPy es la librería fundamental para computación numérica en Python. Todos los frameworks de ML (scikit-learn, PyTorch, TensorFlow) usan NumPy o estructuras compatibles internamente.

**Ventajas clave:**
- Arrays n-dimensionales eficientes (almacenados en memoria contigua)
- Operaciones vectorizadas (ejecutadas en C, no en Python)
- Broadcasting: operar arrays de diferentes formas sin copias
- Ecosistema completo de álgebra lineal, estadística y generación aleatoria

````python
import numpy as np

# --- Creating arrays ---
a = np.array([1, 2, 3, 4, 5])              # From a list
b = np.zeros((3, 4))                        # 3x4 matrix of zeros
c = np.ones((2, 3), dtype=np.float32)       # Specify dtype
d = np.arange(0, 10, 2)                     # Like range(): [0, 2, 4, 6, 8]
e = np.linspace(0, 1, 5)                    # 5 evenly spaced points in [0, 1]
f = np.random.randn(3, 3)                   # Standard normal distribution

print(f"Array a: {a}")
print(f"Shape: {a.shape}, Dtype: {a.dtype}, Ndim: {a.ndim}")
print(f"\nZeros (3x4):\n{b}")
print(f"\nLinspace: {e}")
print(f"\nRandom 3x3:\n{f}")

# --- Reshaping ---
g = np.arange(12)
print(f"\nOriginal: {g} (shape={g.shape})")
print(f"Reshaped to (3,4):\n{g.reshape(3, 4)}")
print(f"Reshaped to (2,2,3):\n{g.reshape(2, 2, 3)}")

# -1 means "infer this dimension"
print(f"Reshape with -1: {g.reshape(3, -1).shape}")  # (3, 4)

# --- Indexing and slicing ---
matrix = np.arange(20).reshape(4, 5)
print(f"\nMatrix:\n{matrix}")
print(f"Row 1: {matrix[1]}")
print(f"Element [2,3]: {matrix[2, 3]}")
print(f"Column 2: {matrix[:, 2]}")
print(f"Submatrix [1:3, 2:4]:\n{matrix[1:3, 2:4]}")

# Boolean indexing (very common in ML)
print(f"Elements > 10: {matrix[matrix > 10]}")
````

## Broadcasting

Broadcasting es el mecanismo que permite a NumPy operar con arrays de diferentes formas sin crear copias en memoria. Es fundamental para escribir código eficiente.

**Reglas de broadcasting:**
1. Si los arrays tienen diferente número de dimensiones, se añaden dimensiones de tamaño 1 al inicio del array más pequeño
2. Arrays con tamaño 1 en una dimensión se comportan como si tuvieran el tamaño del array más grande en esa dimensión
3. Los arrays son compatibles si en cada dimensión tienen el mismo tamaño o uno de ellos tiene tamaño 1

```
Ejemplo: (3,4) + (4,) → (3,4) + (1,4) → (3,4)
         (3,1) + (1,4) → (3,4)
         (3,4) + (3,)  → ERROR (4 != 3)
```

````python
import numpy as np
import time

# --- Broadcasting examples ---
matrix = np.ones((3, 4))
row_vector = np.array([1, 2, 3, 4])          # Shape: (4,)
col_vector = np.array([[10], [20], [30]])     # Shape: (3, 1)

print("Matrix (3,4) + Row (4,):")
print(matrix + row_vector)

print("\nMatrix (3,4) + Column (3,1):")
print(matrix + col_vector)

print("\nRow (4,) + Column (3,1) → (3,4):")
print(row_vector + col_vector)

# Common ML pattern: normalize features (subtract mean, divide by std)
data = np.random.randn(100, 5) * 10 + 50    # 100 samples, 5 features
mean = data.mean(axis=0)                      # Shape: (5,)
std = data.std(axis=0)                        # Shape: (5,)
normalized = (data - mean) / std              # Broadcasting!
print(f"\nNormalized mean: {normalized.mean(axis=0).round(4)}")
print(f"Normalized std: {normalized.std(axis=0).round(4)}")

# --- Vectorization vs loops benchmark ---
n = 1_000_000
a = np.random.randn(n)
b = np.random.randn(n)

# Loop version
start = time.time()
result_loop = np.zeros(n)
for i in range(n):
    result_loop[i] = a[i] * b[i]
loop_time = time.time() - start

# Vectorized version
start = time.time()
result_vec = a * b
vec_time = time.time() - start

print(f"\n--- Benchmark: element-wise multiply ({n:,} elements) ---")
print(f"Loop:       {loop_time:.4f}s")
print(f"Vectorized: {vec_time:.6f}s")
print(f"Speedup:    {loop_time / vec_time:.0f}x")
print(f"Results match: {np.allclose(result_loop, result_vec)}")
````

## Álgebra Lineal con NumPy

El álgebra lineal es el lenguaje del Machine Learning. Operaciones como el producto de matrices, descomposiciones y resolución de sistemas lineales aparecen constantemente.

**Operaciones clave en ML:**
- **Producto punto**: base de las redes neuronales (`y = Wx + b`)
- **Multiplicación de matrices**: propagación de capas
- **Eigenvalues/SVD**: PCA, reducción de dimensionalidad
- **Sistemas lineales**: regresión lineal (solución analítica)

````python
import numpy as np

# --- Dot product ---
# Simulates a single neuron: output = dot(weights, inputs) + bias
weights = np.array([0.5, -0.3, 0.8])
inputs = np.array([1.0, 2.0, 3.0])
bias = 0.1
output = np.dot(weights, inputs) + bias
print(f"Neuron output: {output:.4f}")

# --- Matrix multiplication ---
# Simulates a layer: output = X @ W + b
X = np.random.randn(5, 3)   # 5 samples, 3 features
W = np.random.randn(3, 4)   # 3 inputs → 4 outputs
b = np.random.randn(4)      # 4 biases
layer_output = X @ W + b    # @ is the matrix multiply operator
print(f"\nLayer input shape: {X.shape}")
print(f"Weight matrix shape: {W.shape}")
print(f"Layer output shape: {layer_output.shape}")

# --- Solve a linear system Ax = b ---
# Normal equation for linear regression: (X^T X) w = X^T y
np.random.seed(42)
X_data = np.column_stack([np.ones(50), np.random.randn(50)])  # Add intercept
true_w = np.array([3.0, 1.5])  # True weights
y_data = X_data @ true_w + np.random.randn(50) * 0.5  # Add noise

# Solve using normal equation
w_hat = np.linalg.solve(X_data.T @ X_data, X_data.T @ y_data)
print(f"\n--- Linear Regression via Normal Equation ---")
print(f"True weights:      {true_w}")
print(f"Estimated weights: {w_hat.round(4)}")

# --- Eigenvalues (used in PCA) ---
cov_matrix = np.cov(np.random.randn(3, 100))
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
print(f"\nCovariance matrix shape: {cov_matrix.shape}")
print(f"Eigenvalues: {eigenvalues.round(4)}")
print(f"Variance explained: {(eigenvalues / eigenvalues.sum() * 100).round(2)}%")

# --- SVD (also used in PCA, recommender systems) ---
M = np.random.randn(4, 3)
U, S, Vt = np.linalg.svd(M, full_matrices=False)
print(f"\nSVD shapes: U={U.shape}, S={S.shape}, Vt={Vt.shape}")
print(f"Reconstruction error: {np.linalg.norm(M - U @ np.diag(S) @ Vt):.2e}")
````

## Pandas: Manipulación de datos

Pandas es la herramienta estándar para trabajar con datos tabulares. En un proyecto de ML, pasamos la mayor parte del tiempo cargando, limpiando y transformando datos antes de entrenar modelos.

**Estructuras principales:**
- `Series`: array 1D con índice (como una columna)
- `DataFrame`: tabla 2D con índices de fila y columna

**Operaciones frecuentes en ML:**
- Exploración: `head()`, `describe()`, `info()`, `value_counts()`
- Limpieza: `dropna()`, `fillna()`, `drop_duplicates()`
- Transformación: `apply()`, `map()`, `groupby()`, `merge()`

````python
import pandas as pd
import numpy as np

# --- Create DataFrame from dictionary ---
np.random.seed(42)
n_samples = 200

data = {
    'age': np.random.randint(18, 70, n_samples),
    'salary': np.random.normal(50000, 15000, n_samples).round(2),
    'department': np.random.choice(['Engineering', 'Marketing', 'Sales', 'HR'], n_samples),
    'experience_years': np.random.randint(0, 30, n_samples),
    'satisfaction': np.random.choice([1, 2, 3, 4, 5], n_samples, p=[0.05, 0.1, 0.3, 0.35, 0.2]),
    'left_company': np.random.choice([0, 1], n_samples, p=[0.8, 0.2])
}

df = pd.DataFrame(data)

# Introduce some missing values (realistic scenario)
mask = np.random.random(n_samples) < 0.05
df.loc[mask, 'salary'] = np.nan
df.loc[np.random.random(n_samples) < 0.03, 'satisfaction'] = np.nan

print("=== Dataset Overview ===")
print(f"Shape: {df.shape}")
print(f"\n{df.head()}")
print(f"\n=== Data Types ===")
print(df.dtypes)
print(f"\n=== Missing Values ===")
print(df.isnull().sum())
print(f"\n=== Statistical Summary ===")
print(df.describe().round(2))
print(f"\n=== Target Distribution ===")
print(df['left_company'].value_counts(normalize=True).round(3))
````

````python
import pandas as pd
import numpy as np

# Use the df created in the previous cell

# --- GroupBy operations ---
print("=== Average salary by department ===")
dept_stats = df.groupby('department').agg(
    avg_salary=('salary', 'mean'),
    median_salary=('salary', 'median'),
    count=('salary', 'count'),
    attrition_rate=('left_company', 'mean')
).round(2)
print(dept_stats)

# --- Handling missing values ---
print(f"\n=== Handling Missing Values ===")
print(f"Missing before: {df.isnull().sum().sum()}")

df_clean = df.copy()
# Fill salary with median per department (more robust than global mean)
df_clean['salary'] = df_clean.groupby('department')['salary'].transform(
    lambda x: x.fillna(x.median())
)
# Fill satisfaction with mode
df_clean['satisfaction'] = df_clean['satisfaction'].fillna(df_clean['satisfaction'].mode()[0])
print(f"Missing after:  {df_clean.isnull().sum().sum()}")

# --- Merge example ---
# Create a department info table
dept_info = pd.DataFrame({
    'department': ['Engineering', 'Marketing', 'Sales', 'HR'],
    'budget_millions': [5.0, 3.0, 4.0, 2.0],
    'head': ['Alice', 'Bob', 'Charlie', 'Diana']
})

df_merged = df_clean.merge(dept_info, on='department', how='left')
print(f"\n=== After merge ===")
print(df_merged.head())

# --- Pivot table ---
print(f"\n=== Pivot Table: Avg Salary by Department & Satisfaction ===")
pivot = df_clean.pivot_table(
    values='salary',
    index='department',
    columns='satisfaction',
    aggfunc='mean'
).round(0)
print(pivot)

# --- Feature engineering example ---
df_clean['salary_per_year_exp'] = df_clean['salary'] / (df_clean['experience_years'] + 1)
df_clean['age_group'] = pd.cut(df_clean['age'], bins=[17, 30, 45, 70], labels=['Young', 'Mid', 'Senior'])
print(f"\n=== Engineered Features ===")
print(df_clean[['age', 'age_group', 'salary', 'experience_years', 'salary_per_year_exp']].head())
````

## Visualización con Matplotlib y Seaborn

La visualización de datos es crucial en cada etapa del proceso de ML:
- **Exploración**: entender distribuciones, detectar outliers, ver relaciones
- **Diagnóstico**: evaluar el rendimiento del modelo
- **Comunicación**: presentar resultados

**Matplotlib** ofrece control total, mientras que **Seaborn** proporciona visualizaciones estadísticas con menos código.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Histogram - distribution of salary
axes[0, 0].hist(df_clean['salary'].dropna(), bins=30, edgecolor='black', alpha=0.7, color='steelblue')
axes[0, 0].axvline(df_clean['salary'].mean(), color='red', linestyle='--', label=f"Mean: {df_clean['salary'].mean():,.0f}")
axes[0, 0].set_title('Distribución de Salarios', fontsize=13)
axes[0, 0].set_xlabel('Salario')
axes[0, 0].set_ylabel('Frecuencia')
axes[0, 0].legend()

# 2. Scatter plot - age vs salary colored by department
for dept in df_clean['department'].unique():
    subset = df_clean[df_clean['department'] == dept]
    axes[0, 1].scatter(subset['age'], subset['salary'], alpha=0.5, label=dept, s=30)
axes[0, 1].set_title('Edad vs Salario por Departamento', fontsize=13)
axes[0, 1].set_xlabel('Edad')
axes[0, 1].set_ylabel('Salario')
axes[0, 1].legend(fontsize=9)

# 3. Correlation heatmap
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
corr = df_clean[numeric_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r', center=0,
            square=True, ax=axes[1, 0])
axes[1, 0].set_title('Matriz de Correlación', fontsize=13)

# 4. Box plot - salary by department
df_clean.boxplot(column='salary', by='department', ax=axes[1, 1])
axes[1, 1].set_title('Salario por Departamento', fontsize=13)
axes[1, 1].set_xlabel('Departamento')
axes[1, 1].set_ylabel('Salario')
plt.suptitle('')  # Remove automatic title

plt.tight_layout()
plt.show()

# Bonus: Seaborn pairplot (very useful for EDA)
print("\nPairplot de variables numéricas (primeras 4):")
sample_cols = ['age', 'salary', 'experience_years', 'satisfaction', 'left_company']
sns.pairplot(df_clean[sample_cols].dropna(), hue='left_company', 
             diag_kind='hist', plot_kws={'alpha': 0.5, 's': 20})
plt.suptitle('Pairplot: Variables vs Abandono', y=1.02)
plt.show()
````
